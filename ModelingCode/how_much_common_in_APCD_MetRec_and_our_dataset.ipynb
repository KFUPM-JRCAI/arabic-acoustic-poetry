```python
preprocess = lambda script: araby.strip_tashkeel(araby.strip_tatweel(script))
```

```python
apcd_dataset['clean_bait'] = apcd_dataset['البيت'].apply(lambda script: re.sub('\s\s+',' ',preprocess(script)).strip())
apcd_dataset['clean_bait']
```

# Get MetRec Dataset

```python
metrec_dataset = datasets.load_dataset('metrec')
metrec_dataset
```

```python
def preprocess_example(example):
  example['clean_text'] = re.sub("\s\s+", " ", preprocess(example['text'].replace('#',''))).strip()
  return example
```

```python
metrec_dataset_train = metrec_dataset['train'].map(preprocess_example)
metrec_dataset_test = metrec_dataset['test'].map(preprocess_example)
[item['clean_text'] for item in metrec_dataset_train][:10]
```

```python
dataset_wav_folder = 'dataset_wav'
```

```python
def preprocess_metadata(metadata):
  metadata['Bahr'] = metadata['Bahr'].apply(lambda bahr: bahr.strip() if isinstance(bahr,str) else bahr)
  metadata['Clean Script'] = metadata['Script'].apply(lambda script: re.sub('\s\s+',' ',araby.strip_tashkeel(araby.strip_tatweel(script))).strip())
  metadata['File Exist'] = metadata['Utterance name'].apply(lambda file_name: os.path.isfile(f'{dataset_wav_folder}/{file_name}'))
  metadata = metadata[metadata['File Exist'] == True]
  return metadata
```

```python
all_dataset = preprocess_metadata(pd.read_excel('/content/drive/MyDrive/KFUPM-Master/ICS606/Dataset/metadata.xlsx'))
all_dataset_14_bahr = all_dataset[[bahr not in ('المضارع','المقتضب') for bahr in all_dataset['Bahr']]]
test_dataset = preprocess_metadata(pd.read_csv('/content/drive/MyDrive/KFUPM-Master/ICS606/Dataset/testset.csv'))
```

# our dataset stats

```python
whole_dataset_len = len(all_dataset)
test_dataset_len = len(test_dataset)
print('all dataset rows',whole_dataset_len)
print('test dataset rows',test_dataset_len)
```

# Check how much there in MetRec

```python
whole_dataset_metrec_intersection = len(set(all_dataset['Clean Script'].values).intersection(set([item['clean_text'] for item in metrec_dataset_train])))
test_dataset_metrec_intersection = len(set(test_dataset['Clean Script'].values).intersection(set([item['clean_text'] for item in metrec_dataset_train])))
print('intersection between our whole dataset and metrec',whole_dataset_metrec_intersection,'ratio',(whole_dataset_metrec_intersection/whole_dataset_len)*100)
print('intersection between our test dataset and metrec', test_dataset_metrec_intersection,'ratio',(test_dataset_metrec_intersection/test_dataset_len)*100)
```

# Check how much there in APCD

```python
whole_dataset_apcd_intersection = len(set(all_dataset['Clean Script'].values).intersection(set(apcd_dataset['clean_bait'].values)))
test_dataset_apcd_intersection = len(set(test_dataset['Clean Script'].values).intersection(set(apcd_dataset['clean_bait'].values)))
whole_dataset_apcd_train_intersection = len(set(all_dataset['Clean Script'].values).intersection(set(apcd_dataset_train['clean_bait'].values)))
test_dataset_apcd_train_intersection = len(set(test_dataset['Clean Script'].values).intersection(set(apcd_dataset_train['clean_bait'].values)))
print('intersection between our whole dataset and APCD',whole_dataset_apcd_intersection,'ratio',(whole_dataset_apcd_intersection/whole_dataset_len)*100)
print('intersection between our test dataset and APCD',test_dataset_apcd_intersection,'ratio',(test_dataset_apcd_intersection/test_dataset_len)*100)
print('intersection between our whole dataset and APCD train',whole_dataset_apcd_train_intersection,'ratio',(whole_dataset_apcd_train_intersection/whole_dataset_len)*100)
print('intersection between our test dataset and APCD train',test_dataset_apcd_train_intersection,'ratio',(test_dataset_apcd_train_intersection/test_dataset_len)*100)
```
