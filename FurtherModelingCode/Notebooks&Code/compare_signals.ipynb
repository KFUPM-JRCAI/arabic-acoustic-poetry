```python
#https://stackoverflow.com/questions/18625085/how-to-plot-a-wav-file
def plot_wav_file(wav_file):
  spf = wave.open(wav_file, "r")

  # Extract Raw Audio from Wav File
  signal = spf.readframes(-1)
  signal = np.fromstring(signal, "int16")


  # If Stereo
  if spf.getnchannels() == 2:
      print("Just mono files")
      sys.exit(0)

  plt.figure(1)
  plt.title("Signal Wave...")
  plt.plot(signal)
  plt.show()
```

```python
wav_file = '/content/dataset_wav/P-10.aac'
ipd.display(AudioSegment.from_file(wav_file))
plot_wav_file(wav_file)
```
